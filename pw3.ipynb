1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    sum_even = 0
    for number in numbers:
        if number % 2 == 0:
            sum_even = sum_even + number

    return sum_even
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fib:
    def __init__(self, limit):
        self.limit = limit
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

rslt = Fib(7)

for num in rslt:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
for power in powers_of_two(5):
    print(power)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
C_temps = [0, 20, 37, 100]
F_temps = list(map(lambda c: (c * 9/5) + 32, C_temps))
print("Celsius:", C_temps)
print("Fahrenheit:", F_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    result = filter(lambda char: char not in vowels, input_str)
    return ''.join(result)
user_input = input("Enter a string: ")
no_vowels = remove_vowels(user_input)
print("String without vowels:", no_vowels)
```
